```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#load summary from csv file
summary_df=pd.read_csv('summary_movies.csv')
```

```python
#I created a boxplot as the difference in rating was very important
#for example the 'biography,drama, history' genre may not have had the highest ratings, but because their plot is less spread out it would mean
#its more dependable than 'crime,drama'
plt.figure(figsize=(12,6))
sns.boxplot(x='genre',y='rating', data=summary_df)
plt.title('Box Plot of ratings by Genre')
plt.xlabel('Genre')
plt.ylabel('Ratings')
plt.xticks(rotation=80)
plt.tight_layout()
plt.show()
```
