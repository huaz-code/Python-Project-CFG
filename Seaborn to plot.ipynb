```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#load summary from csv file
summary_df=pd.read_csv('summary_data.csv')
```

```python
#take the 'Total' row out
summary_df=summary_df[summary_df['Month'] != 'Total']
```

```python
#plot the graph using seaborn
sns.barplot(x='Month',y='Monthly Sales', data=summary_df)
plt.title('Monthly Sales for 2018')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

#load summary from csv file
summary_df=pd.read_csv('summary_data.csv')

#take the 'Total' row out
summary_df=summary_df[summary_df['Month'] != 'Total']

#plot the graph using seaborn
sns.barplot(x='Month',y='Monthly Sales', data=summary_df)
plt.title('Monthly Sales for 2018')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.show()
```
